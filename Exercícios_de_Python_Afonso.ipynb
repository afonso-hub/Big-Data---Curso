# 1.12.8 
Uma certa firma fez uma pesquisa de mercado para saber se as pessoas gostaram 
OU não de um novo produto lançado no mercado. Para isso, perfurou em cada cartão o sexo do entrevistado e sua resposta (sim ou não). Sabendo-se que foram entrevistadas 2.000 pessoas, fazer um algoritmo que calcule e escreva:

o número de pessoas que responderam sim; 

o número de pessoas que responderam não; 

a. porcentagem de pessoas do sexo feminino que responderam sim; 
a porcentagem de pessoas do sexo masculino que responderam não

```python
import pandas as pd
from random import random

genero = ['M' if i <=0.5 else 'F' for i in [random() for _ in range(2000)]]
resposta = ['S' if i <=0.5 else 'N' for i in [random() for _ in range(2000)]]


data = pd.DataFrame({'genero':genero, 
                     'resposta':resposta})


data['vol'] = 1

contagem_geral = data.groupby(['genero', 'resposta'])['vol'].count().reset_index()

F = contagem_geral.loc[contagem_geral.genero=='F']
M = contagem_geral.loc[contagem_geral.genero=='M']

S = contagem_geral.loc[contagem_geral.resposta=='S']
N = contagem_geral.loc[contagem_geral.resposta=='N']



total_S = S.groupby('resposta')['vol'].sum().max() 
total_N = N.groupby('resposta')['vol'].sum().max() 

perc_F_S =  F.loc[contagem_geral.resposta=='S'].groupby('resposta')['vol'].sum().max()  / F.groupby('genero')['vol'].sum().max() 
perc_M_N =  M.loc[contagem_geral.resposta=='N'].groupby('resposta')['vol'].sum().max()  / M.groupby('genero')['vol'].sum().max() 



print('o número de pessoas que responderam sim: ', total_S)
print('o número de pessoas que responderam não: ', total_N)
print('a porcentagem de pessoas do sexo feminino que responderam sim: ', perc_F_S)
print('a porcentagem de pessoas do sexo masculino que responderam não: ', perc_M_N)
```

O sistema de avaliação de uma determinada disciplina obedece aos seguintes 
critérios: 
durante o semestre são dadas três notas; 
a nota final é obtida pela média aritmética das notas dadas durante o curso; 
é considerado aprovado o aluno que obtiver a nota final superior ou igual a 60 e que 
tiver comparecido a um mínimo de 40 aulas. 
Fazer um algontmo que: 
a) Leia um conjunto de dados contendo o número de matricula, as três notas e a frequência 
(número de aulas frequentadas) de 100 alunos. 
b) Calcule: 
a nota final de cada aluno; 
a maior e a menor nota da turma; 
a nota média da turma; 
o total de alunos reprovados; 
a porcentagem de alunos reprovados por infrequência. 
c) Escreva: 
para cada aluno, o número de matricula, a frequência, a nota final e o código 
(aprovado ou reprovado); 
o que foi calculado no item b (2, 3, 4 e 5).

```python
import random
import pandas as pd


notas  = [[random.randint(0, 100) for _ in range(3)]  for _ in range(100)]

#como não foi informado o numero maximo de aulas, vou considerar 40 = 75%
#ou seja 54  seria o total(arredondando)

frequencia = [random.randint(0, 54)  for _  in range(100)]


data = pd.DataFrame({'notas':notas, 'frequencia':frequencia})

data.reset_index(inplace=True)


data['media'] = data.apply(lambda row:  sum(row['notas'])/len(row['notas']), axis=1)


print('lista de notas finais')
data.apply(lambda row:  print('matricula: ', row['index'], ',   nota final:', row['media']), axis=1)


data['codigo'] = data.apply(lambda row:  'REPROVADO' if (row['media']<60 or row['frequencia']<40) else 'APROVADO', axis=1)

maior_nota =  data['media'].max()
data['maior nota da turma'] = maior_nota


menor_nota =  data['media'].min()
data['menor nota da turma'] = menor_nota

media_nota_turma =  data['media'].mean()
data['media da nota da turma'] = media_nota_turma


quant_rep_freq = data.loc[(data.media<60) | (data.frequencia<40)].shape[0]
data['total de alunos reprovados']  = quant_rep_freq

percent_rep_freq = data.loc[(data.frequencia<40)].shape[0] / data.shape[0]
data['percent reprovados por frequencia'] = percent_rep_freq




print('')
print('maior nota:', maior_nota)
print('menor nota:', menor_nota)
print('média da turma:', media_nota_turma)
print('total de alunos reprovados:',quant_rep_freq)
print('porcentagem de alunos reprovados pro frequencia:', percent_rep_freq)
```

Tem-se uma estrada ligando várias cidades. Cada cidade tem seu marco quilométrico. Fazer um algoritmo que: 
leia vários pares de dados, contendo cada par os valores dos marcos quilométricos, 
em ordem crescente, de duas cidades. O último par contém estes dois valores iguais; 
Calcule os tempos decomdos para percorrer a distância entre estas duas cidades, 
com as seguintes velocidades: 20, 30, 40, 50, 60, 70 e 80 km/hora, sabendo-se que
t= e/v  = - onde t = tempo; e = espaço; v = velocidade; 

escreva os marcos quilométricos, a velocidade e o tempo decomdo entre as duas 
cidades, apenas quando este tempo for superior a 2 horas.

```python
def calc_tempo_percorrido(marco_cidades, velocidade):

  distancia = marco_cidades[1] - marco_cidades[0]

  tempo = distancia / velocidade

  return tempo

velocidades = list(range(20,90,10))
marcos_cidades = list(zip(range(10,350,50), range(60,400,50)))
marcos_cidades.append((80,80))

cont = 0
while marcos_cidades[cont][0] != marcos_cidades[cont][1]:
  tempo_percorrido = calc_tempo_percorrido(marcos_cidades[cont], velocidades[cont])
  if tempo_percorrido > 2:
    print(marcos_cidades[cont], velocidades[cont], tempo_percorrido)
  cont+=1
```
